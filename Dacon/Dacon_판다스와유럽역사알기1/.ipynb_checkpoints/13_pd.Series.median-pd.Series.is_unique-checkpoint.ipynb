## pandas.Series.median ë©”ì„œë“œ
ì‹œë¦¬ì¦ˆ ë‚´ì˜ ì¤‘ì•™ê°’ì„ ê³„ì‚°í•©ë‹ˆë‹¤.  
ì¤‘ì•™ê°’ì€ ì •ë ¬ëœ ë°ì´í„° ì„¸íŠ¸ì—ì„œ ì¤‘ì•™ì— ìœ„ì¹˜í•œ ê°’ìœ¼ë¡œ, ë°ì´í„° ì„¸íŠ¸ì˜ ì¤‘ì‹¬ ê²½í–¥ì„ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤.

````python
'''
ì˜ˆì œ 1: ì†Œí¬ë¼í…ŒìŠ¤ í•™íŒŒì˜ ë…¼ë¦¬í•™ í…ŒìŠ¤íŠ¸ ì ìˆ˜ ì¤‘ì•™ê°’ ğŸ“šğŸ¤”
ì†Œí¬ë¼í…ŒìŠ¤ í•™íŒŒì—ì„œ ì‹¤ì‹œí•œ ë…¼ë¦¬í•™ í…ŒìŠ¤íŠ¸ì˜ ì ìˆ˜ ì¤‘ì•™ê°’ì„ ê³„ì‚°í•©ë‹ˆë‹¤. ëˆ„ë½ëœ ì ìˆ˜ëŠ” ê³„ì‚°ì—ì„œ ì œì™¸í•©ë‹ˆë‹¤.
'''

import pandas as pd

# ì†Œí¬ë¼í…ŒìŠ¤ í•™íŒŒ í•™ìƒë“¤ì˜ ë…¼ë¦¬í•™ í…ŒìŠ¤íŠ¸ ì ìˆ˜
socratic_logic_scores = pd.Series([85, 90, 95, None, 100, 80, 75, 92])

# ë…¼ë¦¬í•™ í…ŒìŠ¤íŠ¸ ì ìˆ˜ì˜ ì¤‘ì•™ê°’ ê³„ì‚°
median_score = socratic_logic_scores.median(skipna=True)

print(f"ì†Œí¬ë¼í…ŒìŠ¤ í•™íŒŒ ë…¼ë¦¬í•™ í…ŒìŠ¤íŠ¸ ì ìˆ˜ì˜ ì¤‘ì•™ê°’: {median_score}")
````

````python
'''
ì˜ˆì œ 2: í—¬ë ˆë‹ˆì¦˜ ì‹œëŒ€ì˜ ì² í•™ ì„œì  í˜ì´ì§€ ìˆ˜ ì¤‘ì•™ê°’ ğŸ“–
í—¬ë ˆë‹ˆì¦˜ ì‹œëŒ€ì— ì¶œíŒëœ ì² í•™ ì„œì ì˜ í˜ì´ì§€ ìˆ˜ ì¤‘ì•™ê°’ì„ ê³„ì‚°í•©ë‹ˆë‹¤. ì´ëŠ” ë‹¹ì‹œ ì² í•™ ì„œì ì˜ ì¼ë°˜ì ì¸ ë¶„ëŸ‰ì„ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤.
'''

import pandas as pd

# í—¬ë ˆë‹ˆì¦˜ ì‹œëŒ€ ì² í•™ ì„œì ì˜ í˜ì´ì§€ ìˆ˜
hellenistic_philosophy_book_pages = pd.Series([200, 300, 250, 350, 400, 450, 180, 160, 500])

# ì² í•™ ì„œì  í˜ì´ì§€ ìˆ˜ì˜ ì¤‘ì•™ê°’ ê³„ì‚°
median_pages = hellenistic_philosophy_book_pages.median(skipna=True)

print(f"í—¬ë ˆë‹ˆì¦˜ ì‹œëŒ€ ì² í•™ ì„œì  í˜ì´ì§€ ìˆ˜ì˜ ì¤‘ì•™ê°’: {median_pages}")
````

````python
'''
ì˜ˆì œ 2: í”Œë¼í†¤ ì•„ì¹´ë°ë¯¸ì•„ í•™ìƒë“¤ì˜ ìµœì†Œ í•™ìŠµ ì‹œê°„ â³ğŸ“š
í”Œë¼í†¤ ì•„ì¹´ë°ë¯¸ì•„ì— ë‹¤ë‹ˆëŠ” í•™ìƒë“¤ì˜ ì¼ì£¼ì¼ í•™ìŠµ ì‹œê°„ì—ì„œ ìµœì†Œ í•™ìŠµ ì‹œê°„ì„ ì°¾ìŠµë‹ˆë‹¤.
'''

import pandas as pd

# í”Œë¼í†¤ ì•„ì¹´ë°ë¯¸ì•„ í•™ìƒë“¤ì˜ ì¼ì£¼ì¼ í•™ìŠµ ì‹œê°„
academia_study_hours = pd.Series([40, 35, 50, 45, 60, 55, 30, 25])

# ìµœì†Œ í•™ìŠµ ì‹œê°„ ì°¾ê¸°
min_study_hours = academia_study_hours.min(skipna=True)

print(f"í”Œë¼í†¤ ì•„ì¹´ë°ë¯¸ì•„ í•™ìƒë“¤ì˜ ìµœì†Œ ì¼ì£¼ì¼ í•™ìŠµ ì‹œê°„: {min_study_hours}ì‹œê°„")
````

## pandas.Series.mode ë©”ì„œë“œ
ì‹œë¦¬ì¦ˆ ë‚´ì—ì„œ ê°€ì¥ ìì£¼ ë“±ì¥í•˜ëŠ” ê°’ì„ ì°¾ìŠµë‹ˆë‹¤. ì´ ë©”ì„œë“œëŠ” ë²”ì£¼í˜• ë°ì´í„°, ë¬¸ìì—´, ìˆ˜ì¹˜ ë°ì´í„° ë“± ë‹¤ì–‘í•œ ìœ í˜•ì˜ ë°ì´í„°ì— ëŒ€í•´ ìµœë¹ˆê°’ì„ ì°¾ì„ ë•Œ ì‚¬ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
'''
ì˜ˆì œ 1: ì•„í…Œë„¤ ì•„ê³ ë¼ ì‹œì¥ì—ì„œ ê°€ì¥ ì¸ê¸° ìˆëŠ” ìƒí’ˆ ğŸºğŸ‡
ì•„í…Œë„¤ì˜ ì•„ê³ ë¼ ì‹œì¥ì—ì„œ íŒë§¤ë˜ëŠ” ìƒí’ˆ ì¤‘ ê°€ì¥ ì¸ê¸° ìˆëŠ” ìƒí’ˆì„ ì°¾ìŠµë‹ˆë‹¤.
'''

import pandas as pd

# ì•„í…Œë„¤ ì•„ê³ ë¼ ì‹œì¥ì—ì„œ íŒë§¤ë˜ëŠ” ìƒí’ˆ
agora_market_products = pd.Series(['ì˜¬ë¦¬ë¸Œ', 'í¬ë„', 'ì˜¬ë¦¬ë¸Œ', 'ê³¡ë¬¼', 'ì˜¬ë¦¬ë¸Œ', 'í¬ë„', 'ì™€ì¸', 'ì˜¬ë¦¬ë¸Œ'])

# ê°€ì¥ ì¸ê¸° ìˆëŠ” ìƒí’ˆ ì°¾ê¸°
popular_products = agora_market_products.mode(dropna=True)

print(f"ì•„í…Œë„¤ ì•„ê³ ë¼ ì‹œì¥ì—ì„œ ê°€ì¥ ì¸ê¸° ìˆëŠ” ìƒí’ˆ: {popular_products.values}")
````

````python
'''
ì˜ˆì œ 2: ì†Œí”¼ìŠ¤íŠ¸ í•™íŒŒì—ì„œ ê°€ì¥ ìì£¼ ë…¼ì˜ëœ ì£¼ì œ ğŸ“šğŸ—£ï¸
ì†Œí”¼ìŠ¤íŠ¸ í•™íŒŒì—ì„œ ê°€ì¥ ìì£¼ ë…¼ì˜ëœ ì£¼ì œë¥¼ ì°¾ìŠµë‹ˆë‹¤.
'''

import pandas as pd

# ì†Œí”¼ìŠ¤íŠ¸ í•™íŒŒì—ì„œ ë…¼ì˜ëœ ì£¼ì œ
sophist_topics = pd.Series(['ìœ¤ë¦¬', 'ì§„ë¦¬', 'ë„ë•', 'ìœ¤ë¦¬', 'ì§€ì‹', 'ìœ¤ë¦¬', 'ì •ì˜', 'ë„ë•', 'ìœ¤ë¦¬'])

# ê°€ì¥ ìì£¼ ë…¼ì˜ëœ ì£¼ì œ ì°¾ê¸°
frequent_topics = sophist_topics.mode(dropna=True)

print(f"ì†Œí”¼ìŠ¤íŠ¸ í•™íŒŒì—ì„œ ê°€ì¥ ìì£¼ ë…¼ì˜ëœ ì£¼ì œ: {frequent_topics.values}")
````

## pandas.Series.nlargest ë©”ì„œë“œ
ì‹œë¦¬ì¦ˆì—ì„œ ê°€ì¥ í° nê°œì˜ ê°’ì„ ë°˜í™˜í•©ë‹ˆë‹¤. ì´ ë©”ì„œë“œëŠ” í° ê°’ë“¤ì„ ì‹ë³„í•  ë•Œ íŠ¹íˆ ìœ ìš©í•˜ë©°, ìˆ˜ì¹˜í˜• ë°ì´í„°ì— ëŒ€í•´ ë¹ ë¥´ê²Œ ìƒìœ„ nê°œì˜ ê°’ì„ ì°¾ê³ ì í•  ë•Œ ì‚¬ìš©ë©ë‹ˆë‹¤.

````python
'''
ì˜ˆì œ 1: ì˜¬ë¦¼í”¼ì•„ ê²Œì„ì—ì„œì˜ ìµœê³  ê¸°ë¡ ğŸ›ï¸ğŸ…
ì˜¬ë¦¼í”¼ì•„ ê²Œì„ì—ì„œ ì„¸ì›Œì§„ ë‹¤ì–‘í•œ ì¢…ëª©ì˜ ìµœê³  ê¸°ë¡ ì¤‘ ìƒìœ„ 3ê°œë¥¼ ì°¾ìŠµë‹ˆë‹¤.
'''

import pandas as pd

# ì˜¬ë¦¼í”¼ì•„ ê²Œì„ì˜ ì¢…ëª©ë³„ ìµœê³  ê¸°ë¡ (ë‹¨ìœ„: ì´ˆ)
olympia_records = pd.Series([10.32, 9.58, 19.19, 19.32, 43.03, 43.18], 
                            index=['100m', '100m', '200m', '200m', '400m', '400m'])

# ìµœê³  ê¸°ë¡ ìƒìœ„ 3ê°œ ì°¾ê¸°
top_records = olympia_records.nlargest(n=3, keep='all')

print(f"ì˜¬ë¦¼í”¼ì•„ ê²Œì„ ìµœê³  ê¸°ë¡ ìƒìœ„ 3ê°œ: \n{top_records}")
````

````python
'''
ì˜ˆì œ 2: ê³ ëŒ€ ê·¸ë¦¬ìŠ¤ ì² í•™ìë“¤ì˜ ì €ì‘ë¬¼ ìˆ˜ ğŸ“šğŸ¤”
ê³ ëŒ€ ê·¸ë¦¬ìŠ¤ì˜ ìœ ëª…í•œ ì² í•™ìë“¤ì´ ë‚¨ê¸´ ì €ì‘ë¬¼ ìˆ˜ ì¤‘ì—ì„œ ê°€ì¥ ë§ì€ ì €ì‘ë¬¼ì„ ë‚¨ê¸´ ìƒìœ„ 2ëª…ì˜ ì² í•™ìë¥¼ ì°¾ìŠµë‹ˆë‹¤.
'''

import pandas as pd

# ê³ ëŒ€ ê·¸ë¦¬ìŠ¤ ì² í•™ìë“¤ì˜ ì €ì‘ë¬¼ ìˆ˜
philosopher_works = pd.Series([300, 150, 120, 450, 200], 
                              index=['í”Œë¼í†¤', 'ì•„ë¦¬ìŠ¤í† í…”ë ˆìŠ¤', 'ì†Œí¬ë¼í…ŒìŠ¤', 'í”¼íƒ€ê³ ë¼ìŠ¤', 'ì—ìš°ë¦¬í”¼ë°ìŠ¤'])

# ì €ì‘ë¬¼ ìˆ˜ ìƒìœ„ 2ëª…ì˜ ì² í•™ì ì°¾ê¸°
top_philosophers = philosopher_works.nlargest(n=2, keep='all')

print(f"ì €ì‘ë¬¼ì´ ê°€ì¥ ë§ì€ ê³ ëŒ€ ê·¸ë¦¬ìŠ¤ ì² í•™ì ìƒìœ„ 2ëª…: \n{top_philosophers}")
````

## pandas.Series.nsmallest ë©”ì„œë“œ
ì‹œë¦¬ì¦ˆì—ì„œ ê°€ì¥ ì‘ì€ nê°œì˜ ê°’ì„ ë°˜í™˜í•©ë‹ˆë‹¤.  
ì´ ë©”ì„œë“œëŠ” ì‘ì€ ê°’ë“¤ì„ ì‹ë³„í•  ë•Œ íŠ¹íˆ ìœ ìš©í•˜ë©°, ìˆ˜ì¹˜í˜• ë°ì´í„°ì— ëŒ€í•´ ë¹ ë¥´ê²Œ í•˜ìœ„ nê°œì˜ ê°’ì„ ì°¾ê³ ì í•  ë•Œ ì‚¬ìš©ë©ë‹ˆë‹¤.

````python
'''
ì˜ˆì œ 1: ì•„í…Œë„¤ ë§ˆë¼í†¤ì˜ ìµœì†Œ ì™„ì£¼ ì‹œê°„ ğŸƒâ€â™‚ï¸ğŸ†
ì•„í…Œë„¤ ë§ˆë¼í†¤ì—ì„œ ê¸°ë¡ëœ ìµœì†Œ ì™„ì£¼ ì‹œê°„ ì¤‘ í•˜ìœ„ 3ê°œë¥¼ ì°¾ìŠµë‹ˆë‹¤.
'''

import pandas as pd

# ì•„í…Œë„¤ ë§ˆë¼í†¤ì˜ ì™„ì£¼ ì‹œê°„ (ë‹¨ìœ„: ì‹œê°„)
marathon_times = pd.Series([4.5, 3.8, 4.2, 3.9, 5.0, 4.3, 3.7, 4.1])

# ìµœì†Œ ì™„ì£¼ ì‹œê°„ í•˜ìœ„ 3ê°œ ì°¾ê¸°
fastest_times = marathon_times.nsmallest(n=3, keep='all')

print(f"ì•„í…Œë„¤ ë§ˆë¼í†¤ ìµœì†Œ ì™„ì£¼ ì‹œê°„ í•˜ìœ„ 3ê°œ: \n{fastest_times}")
````

````python
'''
ì˜ˆì œ 2: ì•„í…Œë„¤ í•™êµì˜ ìµœì†Œ í•™ê¸‰ ê·œëª¨ ğŸ«ğŸ‘©â€ğŸ«
ì•„í…Œë„¤ì˜ í•™êµì—ì„œ ê° í•™ê¸‰ì˜ í•™ìƒ ìˆ˜ ì¤‘ì—ì„œ ê°€ì¥ ì‘ì€ í•™ê¸‰ ê·œëª¨ í•˜ìœ„ 2ê°œë¥¼ ì°¾ìŠµë‹ˆë‹¤.
'''

import pandas as pd

# ì•„í…Œë„¤ í•™êµì˜ í•™ê¸‰ë³„ í•™ìƒ ìˆ˜
class_sizes = pd.Series([25, 30, 15, 20, 22, 18, 28, 27])

# ìµœì†Œ í•™ê¸‰ ê·œëª¨ í•˜ìœ„ 2ê°œ ì°¾ê¸°
smallest_classes = class_sizes.nsmallest(n=2, keep='all')

print(f"ì•„í…Œë„¤ í•™êµ ìµœì†Œ í•™ê¸‰ ê·œëª¨ í•˜ìœ„ 2ê°œ: \n{smallest_classes}")
````

## pandas.Series.pct_change ë©”ì„œë“œ
ì‹œë¦¬ì¦ˆ ë‚´ì˜ ê° ìš”ì†Œì— ëŒ€í•´ ì´ì „ ìš”ì†Œ ëŒ€ë¹„ ë°±ë¶„ìœ¨ ë³€í™”ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.   
ì´ ë©”ì„œë“œëŠ” ì£¼ë¡œ ê¸ˆìœµ ë°ì´í„° ë¶„ì„ì—ì„œ ì£¼ê°€ì˜ ë°±ë¶„ìœ¨ ë³€í™”ë¥¼ ê³„ì‚°í•˜ëŠ” ë° ì‚¬ìš©ë©ë‹ˆë‹¤.

````python
'''
ì˜ˆì œ 1: ë‚˜ì¼ ê°•ì˜ ì—°ê°„ ë²”ëŒ ë†’ì´ ë³€í™”ìœ¨ ğŸŒŠ
ë‚˜ì¼ ê°•ì˜ ì—°ê°„ ë²”ëŒ ë†’ì´ ë³€í™”ìœ¨ì„ ê³„ì‚°í•˜ì—¬, ë†ì‘ë¬¼ ì¬ë°°ì— ë¯¸ì¹˜ëŠ” ì˜í–¥ì„ ë¶„ì„í•©ë‹ˆë‹¤.
'''

import pandas as pd

# ë‚˜ì¼ ê°•ì˜ ì—°ê°„ ë²”ëŒ ë†’ì´ (ë‹¨ìœ„: ë¯¸í„°)
nile_flood_heights = pd.Series([8.6, 8.9, 9.0, 8.7, 9.2, 9.5, 9.3])

# ì—°ê°„ ë²”ëŒ ë†’ì´ ë³€í™”ìœ¨ ê³„ì‚°
flood_height_changes = nile_flood_heights.pct_change(periods=1)

print(f"ë‚˜ì¼ ê°•ì˜ ì—°ê°„ ë²”ëŒ ë†’ì´ ë³€í™”ìœ¨: \n{flood_height_changes}")
````

````python
'''
ì˜ˆì œ 2: ê³ ëŒ€ ì´ì§‘íŠ¸ ê±´ì¶•ë¬¼ì˜ ì—°ê°„ ê±´ì„¤ ì†ë„ ë³€í™”ìœ¨ ğŸ—ï¸ğŸ“ˆ
ê³ ëŒ€ ì´ì§‘íŠ¸ì—ì„œ ê±´ì„¤ëœ ì£¼ìš” ê±´ì¶•ë¬¼ì˜ ì—°ê°„ ê±´ì„¤ ì†ë„ ë³€í™”ìœ¨ì„ ê³„ì‚°í•©ë‹ˆë‹¤.
'''

import pandas as pd

# ê³ ëŒ€ ì´ì§‘íŠ¸ ê±´ì¶•ë¬¼ì˜ ì—°ê°„ ê±´ì„¤ ì†ë„ (ë‹¨ìœ„: ë¯¸í„°/ë…„)
construction_speeds = pd.Series([20, 25, 22, 30, 28, 32, 31])

# ì—°ê°„ ê±´ì„¤ ì†ë„ ë³€í™”ìœ¨ ê³„ì‚°
speed_changes = construction_speeds.pct_change(periods=1, fill_method='pad')

print(f"ê³ ëŒ€ ì´ì§‘íŠ¸ ê±´ì¶•ë¬¼ì˜ ì—°ê°„ ê±´ì„¤ ì†ë„ ë³€í™”ìœ¨: \n{speed_changes}")
````

## pandas.Series.prod ë©”ì„œë“œ
ì‹œë¦¬ì¦ˆ ë‚´ì˜ ëª¨ë“  ìš”ì†Œì˜ ê³±ì„ ê³„ì‚°í•©ë‹ˆë‹¤.   
ì´ ë©”ì„œë“œëŠ” ìˆ˜ì¹˜ ë°ì´í„°ì— ëŒ€í•œ ê³±ì…ˆ ì§‘ê³„ë¥¼ ìˆ˜í–‰í•  ë•Œ ì‚¬ìš©ë©ë‹ˆë‹¤.

````python
'''
ì˜ˆì œ 1: ì´ì§‘íŠ¸ í”¼ë¼ë¯¸ë“œ ê±´ì„¤ì— ì‚¬ìš©ëœ ëŒ ë¸”ë¡ì˜ ì´ ê³± ğŸºğŸ”¨
ì´ì§‘íŠ¸ í”¼ë¼ë¯¸ë“œ ê±´ì„¤ì— ì‚¬ìš©ëœ ê° ëŒ ë¸”ë¡ì˜ ìˆ˜ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ì‹œë¦¬ì¦ˆì—ì„œ ëª¨ë“  ëŒ ë¸”ë¡ì˜ ì´ ê³±ì„ ê³„ì‚°í•©ë‹ˆë‹¤.
'''

import pandas as pd

# ì´ì§‘íŠ¸ í”¼ë¼ë¯¸ë“œ ê±´ì„¤ì— ì‚¬ìš©ëœ ëŒ ë¸”ë¡ ìˆ˜
pyramid_blocks = pd.Series([10000, 15000, 20000, 25000, 30000])

# ëŒ ë¸”ë¡ì˜ ì´ ê³± ê³„ì‚°
total_blocks_prod = pyramid_blocks.prod(skipna=True)

print(f"ì´ì§‘íŠ¸ í”¼ë¼ë¯¸ë“œ ê±´ì„¤ì— ì‚¬ìš©ëœ ëŒ ë¸”ë¡ì˜ ì´ ê³±: {total_blocks_prod}")
````

````python
'''
ì˜ˆì œ 2: ê³ ëŒ€ ì´ì§‘íŠ¸ ìƒì¸ë“¤ì˜ ê±°ë˜ëŸ‰ ì´ ê³± ğŸ“ˆğŸ›’
ê³ ëŒ€ ì´ì§‘íŠ¸ ì‹œì¥ì—ì„œ ìƒì¸ë“¤ì´ ê±°ë˜í•œ ìƒí’ˆì˜ ìˆ˜ëŸ‰ì„ ë‚˜íƒ€ë‚´ëŠ” ì‹œë¦¬ì¦ˆì—ì„œ ê±°ë˜ëŸ‰ì˜ ì´ ê³±ì„ ê³„ì‚°í•©ë‹ˆë‹¤.
'''

import pandas as pd

# ê³ ëŒ€ ì´ì§‘íŠ¸ ìƒì¸ë“¤ì˜ ê±°ë˜ëŸ‰
merchant_volumes = pd.Series([5, 7, 3, 8, 6, 4, 5])

# ê±°ë˜ëŸ‰ì˜ ì´ ê³± ê³„ì‚°
total_volume_prod = merchant_volumes.prod(skipna=True, min_count=1)

print(f"ê³ ëŒ€ ì´ì§‘íŠ¸ ìƒì¸ë“¤ì˜ ê±°ë˜ëŸ‰ ì´ ê³±: {total_volume_prod}")
````

## pandas.Series.quantile ë©”ì„œë“œ
ì‹œë¦¬ì¦ˆ ë°ì´í„°ì˜ ë¶„ìœ„ìˆ˜ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤. ë¶„ìœ„ìˆ˜ëŠ” ì „ì²´ ë°ì´í„°ë¥¼ ë™ë“±í•œ ë¶€ë¶„ìœ¼ë¡œ ë‚˜ëˆ„ëŠ” ê°’ìœ¼ë¡œ, ì¤‘ì•™ê°’(2ë¶„ìœ„ìˆ˜), ì‚¬ë¶„ìœ„ìˆ˜ ë“±ì„ í¬í•¨í•©ë‹ˆë‹¤.

````python
'''
ì˜ˆì œ 1: ë‚˜ì¼ ê°•ì˜ ì—°ê°„ í™ìˆ˜ ë†’ì´ì˜ ì¤‘ì•™ê°’ê³¼ ì‚¬ë¶„ìœ„ìˆ˜ ğŸŒŠ
ë‚˜ì¼ ê°•ì˜ ì—°ê°„ í™ìˆ˜ ë†’ì´ ë°ì´í„°ì—ì„œ ì¤‘ì•™ê°’ê³¼ í•˜ìœ„ ë° ìƒìœ„ ì‚¬ë¶„ìœ„ìˆ˜ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.
'''

import pandas as pd

# ë‚˜ì¼ ê°•ì˜ ì—°ê°„ í™ìˆ˜ ë†’ì´ (ë‹¨ìœ„: ë¯¸í„°)
nile_flood_heights = pd.Series([8.6, 8.9, 9.0, 8.7, 9.2, 9.5, 9.3])

# ì¤‘ì•™ê°’ê³¼ ì‚¬ë¶„ìœ„ìˆ˜ ê³„ì‚°
quantiles = nile_flood_heights.quantile(q=[0.25, 0.5, 0.75], interpolation='midpoint')

print(f"ë‚˜ì¼ ê°• í™ìˆ˜ ë†’ì´ì˜ ì‚¬ë¶„ìœ„ìˆ˜: \n{quantiles}")
````

````python
'''
ì˜ˆì œ 2: ê³ ëŒ€ ì´ì§‘íŠ¸ ê±´ì¶• í”„ë¡œì íŠ¸ì˜ ê±´ì„¤ ê¸°ê°„ ë¶„ìœ„ìˆ˜ ğŸ—ï¸â³
ê³ ëŒ€ ì´ì§‘íŠ¸ì—ì„œ ì§„í–‰ëœ ì£¼ìš” ê±´ì¶• í”„ë¡œì íŠ¸ì˜ ê±´ì„¤ ê¸°ê°„ ë°ì´í„°ì—ì„œ í•˜ìœ„ 10% ë¶„ìœ„ìˆ˜ì™€ ìƒìœ„ 10% ë¶„ìœ„ìˆ˜ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.
'''

import pandas as pd

# ê³ ëŒ€ ì´ì§‘íŠ¸ ê±´ì¶• í”„ë¡œì íŠ¸ì˜ ê±´ì„¤ ê¸°ê°„ (ë‹¨ìœ„: ë…„)
construction_periods = pd.Series([20, 10, 15, 30, 25, 12, 22])

# í•˜ìœ„ 10%ì™€ ìƒìœ„ 10% ë¶„ìœ„ìˆ˜ ê³„ì‚°
quantiles = construction_periods.quantile(q=[0.1, 0.9], interpolation='nearest')

print(f"ê³ ëŒ€ ì´ì§‘íŠ¸ ê±´ì¶• í”„ë¡œì íŠ¸ ê±´ì„¤ ê¸°ê°„ì˜ ë¶„ìœ„ìˆ˜: \n{quantiles}")
````

## pandas.Series.rank ë©”ì„œë“œ
ì‹œë¦¬ì¦ˆ ë‚´ì˜ ê° ìš”ì†Œì— ìˆœìœ„ë¥¼ í• ë‹¹í•©ë‹ˆë‹¤. ë°ì´í„° ë‚´ì˜ ê° ê°’ì— ëŒ€í•´ ìˆœìœ„ë¥¼ ë§¤ê¸°ë©°, ê°™ì€ ê°’ì— ëŒ€í•´ì„œëŠ” í‰ê·  ìˆœìœ„ë¥¼ í• ë‹¹í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
'''
ì˜ˆì œ 1: ë‚˜ì¼ ê°• í™ìˆ˜ì˜ ì—°ê°„ ë†’ì´ ìˆœìœ„ ğŸŒŠ
ë‚˜ì¼ ê°•ì˜ ì—°ê°„ í™ìˆ˜ ë†’ì´ì— ëŒ€í•´ ìˆœìœ„ë¥¼ ë§¤ê¹ë‹ˆë‹¤. ë™ì¼í•œ í™ìˆ˜ ë†’ì´ì— ëŒ€í•´ì„œëŠ” í‰ê·  ìˆœìœ„ë¥¼ í• ë‹¹í•©ë‹ˆë‹¤.
'''

import pandas as pd

# ë‚˜ì¼ ê°•ì˜ ì—°ê°„ í™ìˆ˜ ë†’ì´ (ë‹¨ìœ„: ë¯¸í„°)
nile_flood_heights = pd.Series([8.6, 8.9, 9.0, 8.7, 9.2, 9.0, 9.3])

# í™ìˆ˜ ë†’ì´ì— ëŒ€í•œ ìˆœìœ„ ë§¤ê¸°ê¸°
flood_height_ranks = nile_flood_heights.rank(method='average', ascending=True)

print(f"ë‚˜ì¼ ê°• í™ìˆ˜ ë†’ì´ì˜ ìˆœìœ„: \n{flood_height_ranks}")
````

````python
'''
ì˜ˆì œ 2: ê³ ëŒ€ ì´ì§‘íŠ¸ ê±´ì¶•ë¬¼ì˜ ìœ„ì—„ ìˆœìœ„ ğŸ›ï¸
ê³ ëŒ€ ì´ì§‘íŠ¸ì˜ ì£¼ìš” ê±´ì¶•ë¬¼ì— ëŒ€í•œ ìœ„ì—„ì„ í‰ê°€í•˜ê³ , ê° ê±´ì¶•ë¬¼ì— ìˆœìœ„ë¥¼ ë§¤ê¹ë‹ˆë‹¤. ë™ì¼í•œ ìœ„ì—„ ì ìˆ˜ë¥¼ ê°€ì§„ ê±´ì¶•ë¬¼ì€ ë¨¼ì € ë‚˜ì˜¨ ê²ƒì´ ë” ë†’ì€ ìˆœìœ„ë¥¼ ë°›ìŠµë‹ˆë‹¤.
'''

import pandas as pd

# ê³ ëŒ€ ì´ì§‘íŠ¸ ê±´ì¶•ë¬¼ì˜ ìœ„ì—„ ì ìˆ˜
monument_grandeur = pd.Series([95, 80, 95, 90, 85, 100], 
                              index=['í”¼ë¼ë¯¸ë“œ', 'ìŠ¤í•‘í¬ìŠ¤', 'ì¹´ë¥´ë‚˜í¬', 'ë£©ì†Œë¥´', 'ì•„ë¶€ì‹¬ë²¨', 'í…Œë² '])

# ìœ„ì—„ ìˆœìœ„ ë§¤ê¸°ê¸°
grandeur_ranks = monument_grandeur.rank(method='first', ascending=False)

print(f"ê³ ëŒ€ ì´ì§‘íŠ¸ ê±´ì¶•ë¬¼ì˜ ìœ„ì—„ ìˆœìœ„: \n{grandeur_ranks}")
````

## pandas.Series.sem ë©”ì„œë“œ
ì‹œë¦¬ì¦ˆì˜ í‘œì¤€ ì˜¤ì°¨ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤. í‘œì¤€ ì˜¤ì°¨ëŠ” ìƒ˜í”Œ ë°ì´í„°ì˜ ë¶„ì‚°ì„ ì¸¡ì •í•˜ëŠ” ê²ƒìœ¼ë¡œ, í‘œë³¸ì˜ í¬ê¸°ê°€ í´ìˆ˜ë¡ í‘œì¤€ ì˜¤ì°¨ëŠ” ì‘ì•„ì§‘ë‹ˆë‹¤.

````python
'''
ì˜ˆì œ 1: ê³ ëŒ€ ì´ì§‘íŠ¸ í•™ìë“¤ì˜ ì‹¤í—˜ ë°ì´í„° í‘œì¤€ ì˜¤ì°¨ ê³„ì‚° ğŸ§ªğŸ“Š
ê³ ëŒ€ ì´ì§‘íŠ¸ í•™ìë“¤ì´ ìˆ˜í–‰í•œ ì‹¤í—˜ì—ì„œ ì–»ì€ ë°ì´í„°ì˜ í‘œì¤€ ì˜¤ì°¨ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.
'''

import pandas as pd

# ê³ ëŒ€ ì´ì§‘íŠ¸ í•™ìë“¤ì˜ ì‹¤í—˜ ë°ì´í„°
experimental_data = pd.Series([3.2, 3.5, 3.8, 3.3, 3.6])

# ì‹¤í—˜ ë°ì´í„°ì˜ í‘œì¤€ ì˜¤ì°¨ ê³„ì‚°
standard_error = experimental_data.sem(ddof=1)

print(f"ê³ ëŒ€ ì´ì§‘íŠ¸ í•™ìë“¤ì˜ ì‹¤í—˜ ë°ì´í„°ì˜ í‘œì¤€ ì˜¤ì°¨: {standard_error}")
````

````python
'''
ì˜ˆì œ 2: ì´ì§‘íŠ¸ ë†ë¶€ë“¤ì˜ ê³¡ë¬¼ ìˆ˜í™•ëŸ‰ í‘œì¤€ ì˜¤ì°¨ ê³„ì‚° ğŸŒ¾ğŸ“‰
ê³ ëŒ€ ì´ì§‘íŠ¸ ë†ë¶€ë“¤ì´ ìˆ˜í™•í•œ ê³¡ë¬¼ì˜ ì–‘ì— ëŒ€í•œ ë°ì´í„°ì˜ í‘œì¤€ ì˜¤ì°¨ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.
'''

import pandas as pd

# ì´ì§‘íŠ¸ ë†ë¶€ë“¤ì˜ ê³¡ë¬¼ ìˆ˜í™•ëŸ‰ ë°ì´í„°
harvest_data = pd.Series([120, 125, 130, 135, 140])

# ê³¡ë¬¼ ìˆ˜í™•ëŸ‰ ë°ì´í„°ì˜ í‘œì¤€ ì˜¤ì°¨ ê³„ì‚°
standard_error = harvest_data.sem(ddof=1)

print(f"ì´ì§‘íŠ¸ ë†ë¶€ë“¤ì˜ ê³¡ë¬¼ ìˆ˜í™•ëŸ‰ ë°ì´í„°ì˜ í‘œì¤€ ì˜¤ì°¨: {standard_error}")
````

## pandas.Series.skew í•¨ìˆ˜

ì£¼ì–´ì§„ Seriesì˜ ë¹„ëŒ€ì¹­ë„(skewness)ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤. ë¹„ëŒ€ì¹­ë„ëŠ” ë°ì´í„° ë¶„í¬ì˜ ë¹„ëŒ€ì¹­ ì •ë„ë¥¼ ë‚˜íƒ€ë‚´ë©°, ì–‘ìˆ˜ì¸ ê²½ìš° ì˜¤ë¥¸ìª½ìœ¼ë¡œ ì¹˜ìš°ì¹œ ë¶„í¬ë¥¼, ìŒìˆ˜ì¸ ê²½ìš° ì™¼ìª½ìœ¼ë¡œ ì¹˜ìš°ì¹œ ë¶„í¬ë¥¼ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤. ì´ í•¨ìˆ˜ì˜ íŒŒë¼ë¯¸í„°ì™€ ë¦¬í„´ê°’ì€ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤:   
   
```íŒŒë¼ë¯¸í„°```:   
- axis: ì—°ì‚°ì„ ìˆ˜í–‰í•  ì¶•ì„ ì§€ì •í•©ë‹ˆë‹¤. ê¸°ë³¸ê°’ì€ 0ìœ¼ë¡œ, ì—´ì„ ë”°ë¼ ê³„ì‚°í•©ë‹ˆë‹¤.
   
```ë¦¬í„´ê°’```:   
- skew: ì£¼ì–´ì§„ Seriesì˜ ë¹„ëŒ€ì¹­ë„ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ë¶€ë™ì†Œìˆ˜ì  ìˆ˜ì¹˜ì…ë‹ˆë‹¤.

````python
# ì½”ë“œ ì˜ˆì œ 1 ğŸ«:

import pandas as pd

# ì´ì§‘íŠ¸ ì‹œëŒ€ ìˆ˜í•™ê³¼ ì² í•™ì„ ê¸°ë°˜ìœ¼ë¡œ í•œ ì½”ë“œ ì˜ˆì œ
# ì‹œëŒ€ì˜ ìˆ«ì ì²´ê³„ë¥¼ ë°˜ì˜í•˜ì—¬ ë°ì´í„° ìƒì„±
data = pd.Series([3, 5, 7, 10, 15, 20])

# ì£¼ì–´ì§„ Seriesì˜ ë¹„ëŒ€ì¹­ë„ ê³„ì‚°
skewness = data.skew(axis=0)

print("ë°ì´í„°ì˜ ë¹„ëŒ€ì¹­ë„:", skewness)
````

## pandas.Series.std í•¨ìˆ˜

## pandas.Series.var í•¨ìˆ˜
ì£¼ì–´ì§„ Seriesì˜ ë¶„ì‚°(variance)ì„ ê³„ì‚°í•©ë‹ˆë‹¤. ë¶„ì‚°ì€ ë°ì´í„°ê°€ í‰ê· ìœ¼ë¡œë¶€í„° ì–¼ë§ˆë‚˜ í¼ì ¸ ìˆëŠ”ì§€ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ì§€í‘œë¡œ, ê°’ì´ ì‘ì„ìˆ˜ë¡ ë°ì´í„°ê°€ í‰ê·  ì£¼ë³€ì— ì§‘ì¤‘ë˜ì–´ ìˆìŒì„ ì˜ë¯¸í•©ë‹ˆë‹¤. ì´ í•¨ìˆ˜ì˜ íŒŒë¼ë¯¸í„°ì™€ ë¦¬í„´ê°’ì€ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤:   
   
```íŒŒë¼ë¯¸í„°```:   
- axis: ì—°ì‚°ì„ ìˆ˜í–‰í•  ì¶•ì„ ì§€ì •í•©ë‹ˆë‹¤. ê¸°ë³¸ê°’ì€ 0ìœ¼ë¡œ, ì—´ì„ ë”°ë¼ ê³„ì‚°í•©ë‹ˆë‹¤. 
- skipna: NA/null ê°’ì„ ì œì™¸í• ì§€ ì—¬ë¶€ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ë¶€ìš¸ ê°’ì…ë‹ˆë‹¤. ê¸°ë³¸ê°’ì€ Trueë¡œ, NA/null ê°’ì„ ì œì™¸í•©ë‹ˆë‹¤.

```ë¦¬í„´ê°’```:
- variance: ì£¼ì–´ì§„ Seriesì˜ ë¶„ì‚°ì„ ë‚˜íƒ€ë‚´ëŠ” ë¶€ë™ì†Œìˆ˜ì  ìˆ˜ì¹˜ì…ë‹ˆë‹¤.

````python
# ì½”ë“œ ì˜ˆì œ 1 ğŸº:

import pandas as pd

# ì´ì§‘íŠ¸ ì‹œëŒ€ ìˆ˜í•™ê³¼ ì² í•™ì„ ê¸°ë°˜ìœ¼ë¡œ í•œ ì½”ë“œ ì˜ˆì œ
# ì‹œëŒ€ì˜ ìˆ«ì ì²´ê³„ë¥¼ ë°˜ì˜í•˜ì—¬ ë°ì´í„° ìƒì„±
data = pd.Series([3, 6, 9, 12, 15])

# ì£¼ì–´ì§„ Seriesì˜ ë¶„ì‚° ê³„ì‚°
variance = data.var(axis=0, skipna=True)

print("ë°ì´í„°ì˜ ë¶„ì‚°:", variance)
````

## pandas.Series.kurtosis í•¨ìˆ˜ëŠ”
ì£¼ì–´ì§„ Seriesì˜ ì²¨ë„(kurtosis)ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.   
ì²¨ë„ëŠ” ë°ì´í„° ë¶„í¬ì˜ ë¾°ì¡±í•œ ì •ë„ë¥¼ ë‚˜íƒ€ë‚´ë©°, ì •ê·œ ë¶„í¬ì™€ ë¹„êµí–ˆì„ ë•Œ ì–¼ë§ˆë‚˜ ì¤‘ì‹¬ì— ëª¨ì—¬ ìˆëŠ”ì§€ë¥¼ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤. ì´ í•¨ìˆ˜ì˜ íŒŒë¼ë¯¸í„°ì™€ ë¦¬í„´ê°’ì€ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤:

```íŒŒë¼ë¯¸í„°```:   
- axis: ì—°ì‚°ì„ ìˆ˜í–‰í•  ì¶•ì„ ì§€ì •í•©ë‹ˆë‹¤. ê¸°ë³¸ê°’ì€ 0ìœ¼ë¡œ, ì—´ì„ ë”°ë¼ ê³„ì‚°í•©ë‹ˆë‹¤.   
   
- skipna: NA/null ê°’ì„ ì œì™¸í• ì§€ ì—¬ë¶€ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ë¶€ìš¸ ê°’ì…ë‹ˆë‹¤. ê¸°ë³¸ê°’ì€ Trueë¡œ, NA/null ê°’ì„ ì œì™¸í•©ë‹ˆë‹¤.   
   
```ë¦¬í„´ê°’```:   
- kurtosis: ì£¼ì–´ì§„ Seriesì˜ ì²¨ë„ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ë¶€ë™ì†Œìˆ˜ì  ìˆ˜ì¹˜ì…ë‹ˆë‹¤.   

````python
# ì½”ë“œ ì˜ˆì œ 1 ğŸ°:

import pandas as pd

# ì¤‘ì„¸ ê³¼í•™ê³¼ ìˆ˜í•™ì„ ê¸°ë°˜ìœ¼ë¡œ í•œ ì½”ë“œ ì˜ˆì œ
# íŠ¹ì • ì‹œëŒ€ì˜ ì¸¡ì • ì²´ê³„ë¥¼ ë°˜ì˜í•˜ì—¬ ë°ì´í„° ìƒì„±
data = pd.Series([1, 2, 3, 4, 5])

# ì£¼ì–´ì§„ Seriesì˜ ì²¨ë„ ê³„ì‚°
kurtosis_value = data.kurtosis(axis=0, skipna=True)

print("ë°ì´í„°ì˜ ì²¨ë„:", kurtosis_value)
````

## pandas.Series.unique í•¨ìˆ˜

counts (ì˜µì…˜): return_counts ë§¤ê°œë³€ìˆ˜ê°€ Trueë¡œ ì„¤ì •ëœ ê²½ìš°, ê° ê³ ìœ í•œ ê°’ì˜ ë“±ì¥ íšŸìˆ˜ë¥¼ í¬í•¨í•˜ëŠ” ë°°ì—´(Array)ì´ ë¦¬í„´ë©ë‹ˆë‹¤.

````python
# ì½”ë“œ ì˜ˆì œ 1 âš¡:
    
import pandas as pd

# ê·¸ë¦¬ìŠ¤ ì‹ í™” ì œìš°ìŠ¤ë¥¼ ê¸°ë°˜ìœ¼ë¡œ í•œ ì½”ë“œ ì˜ˆì œ
# ì œìš°ìŠ¤ì˜ ìì† ì´ë¦„ì„ í¬í•¨í•˜ëŠ” Series ìƒì„±
names = pd.Series(['ì œìš°ìŠ¤', 'í—¤ë¼', 'ì•„í…Œë‚˜', 'í—¤íŒ¨ì´ìŠ¤íŠ¸ìŠ¤', 'ì•„í´ë¡ ', 'ì•„ë¥´í…Œë¯¸ìŠ¤'])

# ê³ ìœ í•œ ê°’ ì°¾ê¸°
unique_names = names.unique()

print("ê³ ìœ í•œ ì´ë¦„ë“¤:", unique_names)
````

````python
# ì½”ë“œ ì˜ˆì œ 2 ğŸŒ©ï¸:

import pandas as pd

# ê·¸ë¦¬ìŠ¤ ì‹ í™” ì œìš°ìŠ¤ë¥¼ ê¸°ë°˜ìœ¼ë¡œ í•œ ì½”ë“œ ì˜ˆì œ
# ì œìš°ìŠ¤ì˜ ìì† ì´ë¦„ê³¼ ë“±ì¥ íšŸìˆ˜ë¥¼ í¬í•¨í•˜ëŠ” Series ìƒì„±
names = pd.Series(['ì œìš°ìŠ¤', 'í—¤ë¼', 'ì•„í…Œë‚˜', 'í—¤íŒ¨ì´ìŠ¤íŠ¸ìŠ¤', 'ì•„í´ë¡ ', 'ì•„ë¥´í…Œë¯¸ìŠ¤', 'í—¤ë¼', 'ì•„í´ë¡ '])

# ê³ ìœ í•œ ê°’ê³¼ ë“±ì¥ íšŸìˆ˜ ì°¾ê¸°
unique_names, counts = names.unique(return_counts=True)

print("ê³ ìœ í•œ ì´ë¦„ë“¤:", unique_names)
print("ì´ë¦„ë³„ ë“±ì¥ íšŸìˆ˜:", counts)
````

```ë¦¬í„´ê°’```:

## pandas.Series.is_unique í•¨ìˆ˜

ì£¼ì–´ì§„ Seriesì˜ ëª¨ë“  ê°’ì´ ê³ ìœ í•œ(unique)í•œì§€ ì—¬ë¶€ë¥¼ íŒë³„í•©ë‹ˆë‹¤.   
ì´ í•¨ìˆ˜ì˜ íŒŒë¼ë¯¸í„°ì™€ ë¦¬í„´ê°’ì€ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤:  
    
```íŒŒë¼ë¯¸í„°```:
ì´ í•¨ìˆ˜ì—ëŠ” íŒŒë¼ë¯¸í„°ê°€ ì—†ìŠµë‹ˆë‹¤.   
   
```ë¦¬í„´ê°’```:

````python
# ì½”ë“œ ì˜ˆì œ 1 âš¡:

import pandas as pd

# ê·¸ë¦¬ìŠ¤ ì‹ í™” ì œìš°ìŠ¤ë¥¼ ê¸°ë°˜ìœ¼ë¡œ í•œ ì½”ë“œ ì˜ˆì œ
# ì œìš°ìŠ¤ì˜ ìì† ì´ë¦„ì„ í¬í•¨í•˜ëŠ” Series ìƒì„±
names = pd.Series(['ì œìš°ìŠ¤', 'í—¤ë¼', 'ì•„í…Œë‚˜', 'í—¤íŒ¨ì´ìŠ¤íŠ¸ìŠ¤', 'ì•„í´ë¡ ', 'ì•„ë¥´í…Œë¯¸ìŠ¤'])

# Seriesì˜ ê°’ì´ ê³ ìœ í•œì§€ í™•ì¸
is_unique = names.is_unique

if is_unique:
    print("ëª¨ë“  ê°’ì´ ê³ ìœ í•©ë‹ˆë‹¤.")
else:
    print("ì¤‘ë³µëœ ê°’ì´ ìˆìŠµë‹ˆë‹¤.")
````
