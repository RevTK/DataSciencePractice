## pd.melt

```python
import pandas as pd

# ë¡œë§ˆ ì œêµ­ì˜ ì£¼ìš” ë„ì‹œë³„ ì¸êµ¬ ë°ì´í„°
data = {
    'City': ['Rome', 'Constantinople', 'Alexandria', 'Antioch'],
    '100 AD': [1000000, 300000, 500000, 150000],
    '200 AD': [1200000, 400000, 600000, 200000],
    '300 AD': [800000, 500000, 750000, 250000]
}
df = pd.DataFrame(data)
print(df)
print(df.info())
print('-----------------------------------------------')

# meltë¥¼ ì‚¬ìš©í•˜ì—¬ ë°ì´í„°ë¥¼ ê¸¸ê²Œ ë³€í™˜(ê¸°ì¡´ 4ì¤„ -> 12ì¤„)
melted_df = pd.melt(df, id_vars=['City'], var_name='Year', value_name='Population')
print(melted_df.head())
print(melted_df.info())
```

## 2. pivot_table

```python
# ë¡œë§ˆ ì œêµ­ì˜ ì „íˆ¬ ë°ì´í„°
battles = {
    'Battle': ['Battle of Actium', 'Battle of Actium', 'Teutoburg Forest', 'Teutoburg Forest'],
    'Year': [-31, -31, 9, 9],
    'Result': ['Victory', 'Defeat', 'Victory', 'Defeat'],
    'Legions Involved': [60, 30, 3, 8]
}
df_battles = pd.DataFrame(battles)
print(df_battles)
print('-----------------------------------------------')

# pivot_tableì„ ì‚¬ìš©í•˜ì—¬ ë°ì´í„° ì§‘ê³„
pivot_battles = df_battles.pivot_table(values='Legions Involved', index='Battle', columns='Result', aggfunc='sum')
print(pivot_battles)
```

```python
# ë¡œë§ˆ ì œêµ­ì˜ ê²½ì œ ì§€í‘œë³„ ì—°ë„ ë°ì´í„°
economy = {
    'Indicator': ['GDP', 'GDP', 'Trade Balance', 'Trade Balance'],
    'Year': ['100 AD', '200 AD', '100 AD', '200 AD'],
    'Value': [50, 70, 10, 15]
}
df_economy = pd.DataFrame(economy)
print(df_economy)
print('-----------------------------------------------')

# pivot_tableì„ ì‚¬ìš©í•˜ì—¬ ë°ì´í„° ì§‘ê³„
pivot_economy = df_economy.pivot_table(values='Value', index='Indicator', columns='Year', aggfunc='sum')
print(pivot_economy)
```

## 6. merge

```python
import pandas as pd

# ìƒ˜í”Œ ë°ì´í„° ìƒì„±
data1 = {
    'ì‹ ': ['ì œìš°ìŠ¤', 'í—¤ë¼', 'í¬ì„¸ì´ëˆ', 'ì•„í…Œë‚˜', 'ì•„í´ë¡œ'],
    'ì†ì„±': ['ì²œë‘¥', 'ê²°í˜¼', 'í•´ì–‘', 'ì§€í˜œ', 'ìŒì•…']
}
data2 = {
    'ì‹ ': ['ì œìš°ìŠ¤', 'í—¤ë¼', 'ì•„í´ë¡œ', 'ì•„ë ˆìŠ¤', 'í—¤ë¥´ë©”ìŠ¤'],
    'ì†Œìœ ë¬¼': ['ë²ˆê°œ', 'ì™•ê´€', 'í™œê³¼ í™”ì‚´', 'ì°½', 'ê¸¸ì¡ì´ì˜ ì§€íŒ¡ì´']
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# 'ì‹ ' ì—´ì„ ê¸°ì¤€ìœ¼ë¡œ ë‘ ë°ì´í„°í”„ë ˆì„ ë³‘í•©
merged_df = pd.merge(df1, df2, on='ì‹ ', how='inner')

print(merged_df)
```

## pandas.concat
- í•˜ë‚˜ ì´ìƒì˜ ë°ì´í„°í”„ë ˆì„ì„ ì—°ê²°(concatenate)í•˜ëŠ” ë° ì‚¬ìš©ë©ë‹ˆë‹¤. 
- ì´ í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ë©´ í–‰ ë˜ëŠ” ì—´ì„ ê¸°ì¤€ìœ¼ë¡œ ë°ì´í„°í”„ë ˆì„ì„ ì—°ê²°í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. 
- ë˜í•œ, ì—°ê²°í•  ë•Œ ë°ì´í„°í”„ë ˆì„ì˜ ì¸ë±ìŠ¤ì™€ ì—´ ì´ë¦„ì„ ìœ ì§€í•˜ê±°ë‚˜ ë‹¤ì‹œ ì§€ì •í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
import pandas as pd

# ìƒ˜í”Œ ë°ì´í„° ìƒì„±
data1 = {
    'ì‚¬ê±´': ['í˜ë¥´ì‹œì•„ ì „ìŸ', 'í ë¡œí¬ë„¤ì†ŒìŠ¤ ì „íˆ¬', 'í…Œë²  ì „ìŸ'],
    'ê²°ê³¼': ['ê·¸ë¦¬ìŠ¤ ìŠ¹ë¦¬', 'ê·¸ë¦¬ìŠ¤ ìŠ¹ë¦¬', 'í…Œë²  ìŠ¹ë¦¬']
}
data2 = {
    'ì‚¬ê±´': ['í”Œë¼íƒ€ì´ì•„ ì „íˆ¬', 'ì‚´ë¼ë¯¸ìŠ¤ ì „íˆ¬'],
    'ê²°ê³¼': ['ê·¸ë¦¬ìŠ¤ ìŠ¹ë¦¬', 'ê·¸ë¦¬ìŠ¤ ìŠ¹ë¦¬']
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# ë°ì´í„°í”„ë ˆì„ ì—°ê²°
concatenated_df = pd.concat([df1, df2], ignore_index=True)

print(concatenated_df)
```
