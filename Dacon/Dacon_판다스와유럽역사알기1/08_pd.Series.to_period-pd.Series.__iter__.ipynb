```python
# ì½”ë“œ ì˜ˆì œ 1: ğŸ›ï¸ ì•„í…Œë„¤ ë¯¼ì£¼ì£¼ì˜ì˜ ì‹œì‘ ğŸ›ï¸
# ì•„í…Œë„¤ëŠ” ê³ ëŒ€ ì„¸ê³„ì—ì„œ ë¯¼ì£¼ì£¼ì˜ë¥¼ ì²˜ìŒ ì‹¤í˜„í•œ ë„ì‹œêµ­ê°€ì…ë‹ˆë‹¤. 
# ì´ ì½”ë“œ ì˜ˆì œì—ì„œëŠ” ì•„í…Œë„¤ ë¯¼ì£¼ì£¼ì˜ì˜ ì£¼ìš” ì‚¬ê±´ë“¤ê³¼ í•´ë‹¹ ì‚¬ê±´ë“¤ì˜ ì—°ë„ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³ , 
# to_period ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ì—°ê°„ ê¸°ê°„ìœ¼ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.

import pandas as pd

# ì•„í…Œë„¤ ë¯¼ì£¼ì£¼ì˜ì˜ ì£¼ìš” ì‚¬ê±´ë“¤ê³¼ ì—°ë„
democracy_events = pd.Series(["Reforms of Cleisthenes", "Persian Wars", "Peloponnesian War"],
                             index=pd.to_datetime(["1908-01-01", "1890-01-01", "1931-01-01"]))
#to_datetime(["508-01-01", "490-01-01", "431-01-01"]))

# ì—°ê°„ ê¸°ê°„ìœ¼ë¡œ ë³€í™˜
events_period = democracy_events.to_period("Y")

print("ğŸ›ï¸ ì•„í…Œë„¤ ë¯¼ì£¼ì£¼ì˜ì˜ ì£¼ìš” ì‚¬ê±´ ê¸°ê°„:", events_period)
```

```python
# ì½”ë“œ ì˜ˆì œ 2: ğŸ­ ì•„í…Œë„¤ì˜ ê³ ì „ê¸° ğŸ­
# ì•„í…Œë„¤ì˜ ê³ ì „ê¸°ëŠ” ë¬¸í™”ì™€ ì˜ˆìˆ ì´ ê½ƒí”¼ìš´ ì‹œê¸°ë¡œ, ë§ì€ ê³ ì „ ë¬¸í•™ê³¼ ê·¹ì‘í’ˆì´ íƒ„ìƒí–ˆìŠµë‹ˆë‹¤.
# ì´ ì½”ë“œ ì˜ˆì œì—ì„œëŠ” ê³ ì „ê¸° ì£¼ìš” ê·¹ì‘í’ˆë“¤ê³¼ ê·¸ ì‘í’ˆë“¤ì˜ ì´ˆì—° ì—°ë„ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³ , 
# to_period ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ì—°ê°„ ê¸°ê°„ìœ¼ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.

import pandas as pd

# ê³ ì „ê¸° ì£¼ìš” ê·¹ì‘í’ˆë“¤ê³¼ ì´ˆì—° ì—°ë„
classic_plays = pd.Series(["Oresteia", "Oedipus Rex", "Medea"],
                          index=pd.to_datetime(["1858-01-01", "1829-01-01", "1831-01-01"]))

#"458-01-01", "429-01-01", "431-01-01"

# ì—°ê°„ ê¸°ê°„ìœ¼ë¡œ ë³€í™˜
plays_period = classic_plays.to_period("Y")

print("ğŸ­ ì•„í…Œë„¤ ê³ ì „ê¸°ì˜ ì£¼ìš” ê·¹ì‘í’ˆ ê¸°ê°„:", plays_period)
```

```python
# ì½”ë“œ ì˜ˆì œ 3: âš”ï¸ ì•„í…Œë„¤ì˜ ì „ìŸë“¤ âš”ï¸
# ì•„í…Œë„¤ëŠ” ì—¬ëŸ¬ ì¤‘ìš”í•œ ì „ìŸì— ì°¸ì—¬í–ˆìŠµë‹ˆë‹¤. 
# ì´ ì½”ë“œ ì˜ˆì œì—ì„œëŠ” ì•„í…Œë„¤ê°€ ì°¸ì—¬í•œ ì£¼ìš” ì „ìŸë“¤ê³¼ ê·¸ ì „ìŸë“¤ì˜ ë°œë°œ ì—°ë„ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³ , 
# to_period ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ì—°ê°„ ê¸°ê°„ìœ¼ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.

import pandas as pd

# ì•„í…Œë„¤ê°€ ì°¸ì—¬í•œ ì£¼ìš” ì „ìŸë“¤ê³¼ ë°œë°œ ì—°ë„
wars = pd.Series(["Greco-Persian Wars", "Peloponnesian War", "Corinthian War"],
                 index=pd.to_datetime(["1899-01-01", "1831-01-01", "1795-01-01"]))


#"499-01-01", "431-01-01", "395-01-01"
# ì—°ê°„ ê¸°ê°„ìœ¼ë¡œ ë³€í™˜
wars_period = wars.to_period("Y")

print("âš”ï¸ ì•„í…Œë„¤ì˜ ì£¼ìš” ì „ìŸ ê¸°ê°„:", wars_period)
```

## pandas.Series.to_timestamp ë©”ì„œë“œ

```python
# ì½”ë“œ ì˜ˆì œ 1: ğŸº ì•„í…Œë„¤ì˜ ê³ ëŒ€ ì˜¬ë¦¼í”½ ğŸº
# ì•„í…Œë„¤ëŠ” ê³ ëŒ€ ì˜¬ë¦¼í”½ì— ì ê·¹ì ìœ¼ë¡œ ì°¸ì—¬í–ˆìŠµë‹ˆë‹¤. 
# ì´ ì½”ë“œ ì˜ˆì œì—ì„œëŠ” ì•„í…Œë„¤ ì„ ìˆ˜ë“¤ì´ ì°¸ê°€í•œ ê³ ëŒ€ ì˜¬ë¦¼í”½ì˜ ê¸°ê°„ì„ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³ , 
# to_timestamp ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ì‹œì‘ íƒ€ì„ìŠ¤íƒ¬í”„ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.

import pandas as pd

# ì•„í…Œë„¤ ì„ ìˆ˜ë“¤ì´ ì°¸ê°€í•œ ê³ ëŒ€ ì˜¬ë¦¼í”½ ê¸°ê°„
olympics = pd.Series(["2076", "1772", "1744"], name="Athenian Olympics Participation")
#"776 BC", "472 BC", "444 BC"
olympics_period = pd.PeriodIndex(olympics, freq="Y")

# ì‹œì‘ íƒ€ì„ìŠ¤íƒ¬í”„ë¡œ ë³€í™˜
olympics_timestamp = olympics_period.to_timestamp()

print("ğŸº ì•„í…Œë„¤ì˜ ê³ ëŒ€ ì˜¬ë¦¼í”½ ì°¸ê°€ ì‹œì‘ íƒ€ì„ìŠ¤íƒ¬í”„:", olympics_timestamp)
```

```python
# ì½”ë“œ ì˜ˆì œ 2: ğŸ“œ ì•„í…Œë„¤ì˜ ë²•ë ¹ ğŸ“œ
# ì•„í…Œë„¤ëŠ” ë‹¤ì–‘í•œ ì¤‘ìš”í•œ ë²•ë ¹ì„ ì œì •í–ˆìŠµë‹ˆë‹¤. 
# ì´ ì½”ë“œ ì˜ˆì œì—ì„œëŠ” ëª‡ëª‡ ì£¼ìš” ì•„í…Œë„¤ ë²•ë ¹ì˜ ì‹œí–‰ ê¸°ê°„ì„ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³ , 
# to_timestamp ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ê·¸ ê¸°ê°„ì˜ ì‹œì‘ íƒ€ì„ìŠ¤íƒ¬í”„ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.

import pandas as pd

# ì£¼ìš” ì•„í…Œë„¤ ë²•ë ¹ì˜ ì‹œí–‰ ê¸°ê°„
laws = pd.Series(["Solon's Reforms", "Cleisthenes' Reforms", "Pericles' Citizenship Law"],
                 index=pd.PeriodIndex(["2076", "1772", "1744"], freq="Y")) 
#["594 BC", "508 BC", "451 BC"]

# ì‹œì‘ íƒ€ì„ìŠ¤íƒ¬í”„ë¡œ ë³€í™˜
laws_timestamp = laws.index.to_timestamp()

print("ğŸ“œ ì•„í…Œë„¤ì˜ ì£¼ìš” ë²•ë ¹ ì‹œí–‰ ì‹œì‘ íƒ€ì„ìŠ¤íƒ¬í”„:", laws_timestamp)
```

```python
ì½”ë“œ ì˜ˆì œ 3: âš”ï¸ ì•„í…Œë„¤ ì „ìŸ ê¸°ê°„ âš”ï¸
ì•„í…Œë„¤ëŠ” ì—¬ëŸ¬ ì „ìŸì— ì°¸ì—¬í–ˆìŠµë‹ˆë‹¤.
ì´ ì½”ë“œ ì˜ˆì œì—ì„œëŠ” ì•„í…Œë„¤ê°€ ì°¸ì—¬í•œ ì£¼ìš” ì „ìŸì˜ ê¸°ê°„ì„ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³ ,
to_timestamp ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ê° ì „ìŸì˜ ì‹œì‘ íƒ€ì„ìŠ¤íƒ¬í”„ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.

import pandas as pd

# ì•„í…Œë„¤ê°€ ì°¸ì—¬í•œ ì£¼ìš” ì „ìŸì˜ ê¸°ê°„
wars = pd.Series(["Persian Wars", "Peloponnesian War", "Corinthian War"],
                 index=pd.PeriodIndex(["2076", "1772", "1744"], freq="Y"))
#["499 BC-449 BC", "431 BC-404 BC", "395 BC-387 BC"]
# ì‹œì‘ íƒ€ì„ìŠ¤íƒ¬í”„ë¡œ ë³€í™˜
wars_timestamp = wars.index.to_timestamp(how="start")

print("âš”ï¸ ì•„í…Œë„¤ ì „ìŸ ê¸°ê°„ì˜ ì‹œì‘ íƒ€ì„ìŠ¤íƒ¬í”„:", wars_timestamp)
```
