## pandas.Series.sub ë©”ì„œë“œ

```python
# ì½”ë“œ ì˜ˆì œ 1: ğŸº íˆíƒ€ì´íŠ¸ ì œêµ­ì˜ ì˜í†  ë³€í™” ğŸº
# íˆíƒ€ì´íŠ¸ ì œêµ­ì˜ ì˜í† ëŠ” ì—­ì‚¬ì ìœ¼ë¡œ ì—¬ëŸ¬ ë²ˆì˜ ë³€í™”ë¥¼ ê²ªì—ˆìŠµë‹ˆë‹¤. 
# ì´ ì½”ë“œ ì˜ˆì œì—ì„œëŠ” íŠ¹ì • ì‹œê¸°ì˜ ì˜í†  ë©´ì ê³¼ ê·¸ ë‹¤ìŒ ì‹œê¸°ì˜ ì˜í†  ë©´ì ì„ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³ , 
# sub ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ì˜í†  ë©´ì ì˜ ë³€í™”ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.

import pandas as pd

# íŠ¹ì • ì‹œê¸°ì˜ íˆíƒ€ì´íŠ¸ ì œêµ­ ì˜í†  ë©´ì  (ë‹¨ìœ„: ë§Œ kmÂ²)
territory_initial = pd.Series({"ì¤‘ì•™ ì•„ë‚˜í†¨ë¦¬ì•„": 30, "ë¶ë¶€ ì‹œë¦¬ì•„": 20}, name="Initial Territory")

# ê·¸ ë‹¤ìŒ ì‹œê¸°ì˜ ì˜í†  ë©´ì  (ë‹¨ìœ„: ë§Œ kmÂ²)
territory_later = pd.Series({"ì¤‘ì•™ ì•„ë‚˜í†¨ë¦¬ì•„": 25, "ë¶ë¶€ ì‹œë¦¬ì•„": 15, "ë™ë¶€ ì§€ì¤‘í•´": 10}, name="Later Territory")

# ì˜í†  ë©´ì ì˜ ë³€í™” ê³„ì‚°
territory_change = territory_initial.sub(territory_later, fill_value=0)

print("ğŸº íˆíƒ€ì´íŠ¸ ì œêµ­ì˜ ì˜í†  ë³€í™” (ë‹¨ìœ„: ë§Œ kmÂ²):")
print(territory_change)
```

## pandas.Series.sub

```python
# ì½”ë“œ ì˜ˆì œ 1: ğŸ° ì•„ì‹œë¦¬ì•„ ì œêµ­ì˜ ì˜í†  í™•ì¥ ğŸ°
# ì•„ì‹œë¦¬ì•„ ì œêµ­ì€ ê°•ë ¥í•œ êµ°ì‚¬ë ¥ìœ¼ë¡œ ì—¬ëŸ¬ ì§€ì—­ì„ ì •ë³µí•˜ì—¬ ì˜í† ë¥¼ í™•ì¥í–ˆìŠµë‹ˆë‹¤.
# ì´ ì½”ë“œ ì˜ˆì œì—ì„œëŠ” ì•„ì‹œë¦¬ì•„ ì œêµ­ì˜ ì˜í† ê°€ í™•ì¥ë˜ê¸° ì „í›„ì˜ ë©´ì ì„ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³ , 
# sub ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ì˜í†  í™•ì¥ ì „í›„ì˜ ë©´ì  ì°¨ì´ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤. 
# ì´ë•Œ, fill_value íŒŒë¼ë¯¸í„°ë¥¼ ì‚¬ìš©í•˜ì—¬ í™•ì¥ë˜ê¸° ì „ì—ëŠ” ì¡´ì¬í•˜ì§€ ì•Šì•˜ë˜ ì˜í† ì˜ ë©´ì ì„ 0ìœ¼ë¡œ ê°„ì£¼í•©ë‹ˆë‹¤.

import pandas as pd

# ì•„ì‹œë¦¬ì•„ ì œêµ­ì˜ ì˜í†  í™•ì¥ ì „ ë©´ì  (ë‹¨ìœ„: ë§Œ kmÂ²)
territory_before = pd.Series({"ì¤‘ì•™ ì•„ì‹œë¦¬ì•„": 15, "ë¶ë¶€ ë©”ì†Œí¬íƒ€ë¯¸ì•„": 10}, name="Territory Before Expansion")

# ì˜í†  í™•ì¥ í›„ ë©´ì  (ë‹¨ìœ„: ë§Œ kmÂ²)
territory_after = pd.Series({"ì¤‘ì•™ ì•„ì‹œë¦¬ì•„": 20, "ë¶ë¶€ ë©”ì†Œí¬íƒ€ë¯¸ì•„": 15, "ë‚¨ë¶€ ë©”ì†Œí¬íƒ€ë¯¸ì•„": 10}, name="Territory After Expansion")

# ì˜í†  í™•ì¥ì— ë”°ë¥¸ ë©´ì  ì°¨ì´ ê³„ì‚°
territory_expansion = territory_after.sub(territory_before, fill_value=0)

print("ğŸ° ì•„ì‹œë¦¬ì•„ ì œêµ­ì˜ ì˜í†  í™•ì¥ ê²°ê³¼ (ë‹¨ìœ„: ë§Œ kmÂ²):")
print(territory_expansion)
```

```python
# ì½”ë“œ ì˜ˆì œ 2-1: ğŸŒ¾ ì•„ì‹œë¦¬ì•„ ì œêµ­ì˜ ë†ì—… ìƒì‚° ì¦ê°€ ğŸŒ¾
# ì•„ì‹œë¦¬ì•„ ì œêµ­ì€ ê°•ë ¥í•œ ë†ì—… ê¸°ë°˜ì„ ê°€ì§€ê³  ìˆì—ˆìœ¼ë©°, íŠ¹ì • ì‹œê¸°ì— ìƒì‚°ëŸ‰ì´ ì¦ê°€í–ˆìŠµë‹ˆë‹¤. 
# ì´ ì½”ë“œ ì˜ˆì œì—ì„œëŠ” ì´ˆê¸° ë†ì—… ìƒì‚°ëŸ‰ê³¼ ì¦ê°€ìœ¨ì„ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³ , 
# mul ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ì¦ê°€ í›„ì˜ ìƒì‚°ëŸ‰ì„ ê³„ì‚°í•©ë‹ˆë‹¤.

import pandas as pd

# ì•„ì‹œë¦¬ì•„ ì œêµ­ì˜ ì´ˆê¸° ë†ì—… ìƒì‚°ëŸ‰ (ë‹¨ìœ„: í†¤)
initial_production = pd.Series({"ë°€": 1000, "ë³´ë¦¬": 1500}, name="Initial Production")

# ìƒì‚° ì¦ê°€ìœ¨
increase_rate = pd.Series({"ë°€": 1.2, "ë³´ë¦¬": 1.1}, name="Increase Rate")

# ì¦ê°€ í›„ì˜ ìƒì‚°ëŸ‰ ê³„ì‚°
increased_production = initial_production.mul(increase_rate)

print("ğŸŒ¾ ì•„ì‹œë¦¬ì•„ ì œêµ­ì˜ ì¦ê°€ í›„ ë†ì—… ìƒì‚°ëŸ‰ (ë‹¨ìœ„: í†¤):")
print(increased_production)
```

```python
# ì½”ë“œ ì˜ˆì œ 2-2: ğŸ’° ì•„ì‹œë¦¬ì•„ ì œêµ­ì˜ êµì—­ ê·œëª¨ í™•ëŒ€ ğŸ’°
# ì•„ì‹œë¦¬ì•„ ì œêµ­ì€ ê´‘ë²”ìœ„í•œ êµì—­ ë„¤íŠ¸ì›Œí¬ë¥¼ í†µí•´ ë¶€ë¥¼ ì¶•ì í–ˆìŠµë‹ˆë‹¤. 
# ì´ ì½”ë“œ ì˜ˆì œì—ì„œëŠ” ì´ˆê¸° êµì—­ ê·œëª¨ì™€ êµì—­ í™•ëŒ€ìœ¨ì„ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³ ,
# mul ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ í™•ëŒ€ í›„ì˜ êµì—­ ê·œëª¨ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.

import pandas as pd

# ì•„ì‹œë¦¬ì•„ ì œêµ­ì˜ ì´ˆê¸° êµì—­ ê·œëª¨ (ë‹¨ìœ„: ì²œ ê¸ˆ)
initial_trade = pd.Series({"ê¸ˆ": 500, "ì€": 800}, name="Initial Trade")

# êµì—­ í™•ëŒ€ìœ¨
trade_expansion = pd.Series({"ê¸ˆ": 1.3, "ì€": 1.2}, name="Trade Expansion")

# í™•ëŒ€ í›„ì˜ êµì—­ ê·œëª¨ ê³„ì‚°
expanded_trade = initial_trade.mul(trade_expansion, fill_value=1)

print("ğŸ’° ì•„ì‹œë¦¬ì•„ ì œêµ­ì˜ í™•ëŒ€ í›„ êµì—­ ê·œëª¨ (ë‹¨ìœ„: ì²œ ê¸ˆ):")
print(expanded_trade)
```

## pandas.Series.div ë©”ì„œë“œëŠ”

- ì´ ë©”ì„œë“œëŠ” ë¶„ì„ ì‘ì—…ì—ì„œ ë¹„ìœ¨ì´ë‚˜ í‰ê· ì„ ê³„ì‚°í•  ë•Œ íŠ¹íˆ ìœ ìš©í•©ë‹ˆë‹¤.

```python
# ì½”ë“œ ì˜ˆì œ 1: ğŸ› ì•„ì‹œë¦¬ì•„ ì œêµ­ì˜ ì¸êµ¬ ë°€ë„ ğŸ›
# ì•„ì‹œë¦¬ì•„ ì œêµ­ì€ ê´‘ëŒ€í•œ ì˜í† ë¥¼ ì§€ë°°í–ˆìœ¼ë©°, ì´ ì½”ë“œ ì˜ˆì œì—ì„œëŠ” ì•„ì‹œë¦¬ì•„ ì œêµ­ì˜ íŠ¹ì • ì§€ì—­ ì¸êµ¬ì™€ 
# í•´ë‹¹ ì§€ì—­ ë©´ì ì„ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³ , 
# div ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ì¸êµ¬ ë°€ë„ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.

import pandas as pd

# ì•„ì‹œë¦¬ì•„ ì œêµ­ì˜ íŠ¹ì • ì§€ì—­ ì¸êµ¬ (ë‹¨ìœ„: ë§Œ ëª…)
population = pd.Series({"ë‹ˆë„¤ë² ": 120, "ì•„ìŠˆë¥´": 80, "ì¹¼í›„": 60}, name="Population")

# í•´ë‹¹ ì§€ì—­ì˜ ë©´ì  (ë‹¨ìœ„: kmÂ²)
area = pd.Series({"ë‹ˆë„¤ë² ": 50, "ì•„ìŠˆë¥´": 30, "ì¹¼í›„": 20}, name="Area")

# ì¸êµ¬ ë°€ë„ ê³„ì‚° (ë‹¨ìœ„: ëª…/kmÂ²)
population_density = population.div(area)

print("ğŸ› ì•„ì‹œë¦¬ì•„ ì œêµ­ì˜ ì¸êµ¬ ë°€ë„ (ë‹¨ìœ„: ëª…/kmÂ²):")
print(population_density)
```

```python
# ì½”ë“œ ì˜ˆì œ 2: ğŸ’° ì•„ì‹œë¦¬ì•„ ì œêµ­ì˜ ì¬í™” ë¶„ë°° ğŸ’°
# ì•„ì‹œë¦¬ì•„ ì œêµ­ì€ ì „ë¦¬í’ˆê³¼ ìì›ì„ ë¶„ë°°í•˜ëŠ” ë³µì¡í•œ ì‹œìŠ¤í…œì„ ê°€ì§€ê³  ìˆì—ˆìŠµë‹ˆë‹¤. 
# ì´ ì½”ë“œ ì˜ˆì œì—ì„œëŠ” ì „ë¦¬í’ˆì˜ ì´ëŸ‰ê³¼ ë¶„ë°°í•´ì•¼ í•  ë‹¨ìœ„ì˜ ìˆ˜ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³ , 
# div ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ê° ë‹¨ìœ„ì— ë¶„ë°°ë  ì „ë¦¬í’ˆì˜ ì–‘ì„ ê³„ì‚°í•©ë‹ˆë‹¤.

import pandas as pd

# ì „ë¦¬í’ˆì˜ ì´ëŸ‰ (ë‹¨ìœ„: ì²œ ê¸ˆ)
spoils = pd.Series({"ê¸ˆ": 100, "ì€": 200, "ë™": 300}, name="Spoils")

# ë¶„ë°°í•´ì•¼ í•  ë‹¨ìœ„ì˜ ìˆ˜
units = pd.Series({"ê¸ˆ": 4, "ì€": 5, "ë™": 6}, name="Units")

# ê° ë‹¨ìœ„ì— ë¶„ë°°ë  ì „ë¦¬í’ˆì˜ ì–‘ ê³„ì‚° (ë‹¨ìœ„: ì²œ ê¸ˆ)
spoils_per_unit = spoils.div(units, fill_value=1)

print("ğŸ’° ì•„ì‹œë¦¬ì•„ ì œêµ­ì˜ ê° ë‹¨ìœ„ì— ë¶„ë°°ë  ì „ë¦¬í’ˆì˜ ì–‘ (ë‹¨ìœ„: ì²œ ê¸ˆ):")
print(spoils_per_unit)
```

## pandas.Series.truediv ë©”ì„œë“œ
- Series ê°ì²´ì˜ ê° ìš”ì†Œë¥¼ ë‹¤ë¥¸ Series ê°ì²´ë‚˜ ìŠ¤ì¹¼ë¼ ê°’ìœ¼ë¡œ ë‚˜ëˆ„ëŠ” ì—°ì‚°ì„ ìˆ˜í–‰í•˜ë©°,í•­ìƒ ë¶€ë™ ì†Œìˆ˜ì  ê²°ê³¼ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤. ì´ ë©”ì„œë“œëŠ” Python 3ì˜ íŠ¹ì„±ì¸ ì§„ì •í•œ ë‚˜ëˆ—ì…ˆ(/)ê³¼ ë™ì¼í•˜ê²Œ ì‘ë™í•©ë‹ˆë‹¤.

```python
# ì½”ë“œ ì˜ˆì œ 1: ğŸŒ¾ ì•„ì‹œë¦¬ì•„ ì œêµ­ì˜ ì‹ëŸ‰ ë°°ë¶„ ğŸŒ¾
# ì•„ì‹œë¦¬ì•„ ì œêµ­ì€ ì¤‘ì•™ ì§‘ì¤‘ì‹ ì‹ëŸ‰ ì €ì¥ ë° ë°°ë¶„ ì‹œìŠ¤í…œì„ ê°€ì§€ê³  ìˆì—ˆìŠµë‹ˆë‹¤. 
# ì´ ì½”ë“œ ì˜ˆì œì—ì„œëŠ” ì €ì¥ëœ ì‹ëŸ‰ì˜ ì–‘ê³¼ ë°°ë¶„í•´ì•¼ í•  êµ°ëŒ€ì˜ ìˆ˜ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³ , 
# truediv ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ êµ°ëŒ€ ë‹¹ ë°°ë¶„ë˜ëŠ” ì‹ëŸ‰ì˜ ì–‘ì„ ê³„ì‚°í•©ë‹ˆë‹¤.

import pandas as pd


# ì €ì¥ëœ ì‹ëŸ‰ì˜ ì–‘ (ë‹¨ìœ„: í†¤)
stored_food = pd.Series({"ê³¡ë¬¼": 1000, "ì•¼ì±„": 500}, name="Stored Food")

# ë°°ë¶„í•´ì•¼ í•  êµ°ëŒ€ì˜ ìˆ˜
army_units = pd.Series({"ê³¡ë¬¼": 10, "ì•¼ì±„": 5}, name="Army Units")

# êµ°ëŒ€ ë‹¹ ë°°ë¶„ë˜ëŠ” ì‹ëŸ‰ì˜ ì–‘ ê³„ì‚° (ë‹¨ìœ„: í†¤)
food_per_unit = stored_food.truediv(army_units)

print("ğŸŒ¾ êµ°ëŒ€ ë‹¹ ë°°ë¶„ë˜ëŠ” ì‹ëŸ‰ì˜ ì–‘ (ë‹¨ìœ„: í†¤):")
print(food_per_unit)
```

```python
# ì½”ë“œ ì˜ˆì œ 2: ğŸ’§ ì•„ì‹œë¦¬ì•„ì˜ ë¬¼ ìì› ë¶„ë°° ğŸ’§
# ì•„ì‹œë¦¬ì•„ ì œêµ­ì€ ë¬¼ ìì› ê´€ë¦¬ì— ìˆì–´ì„œë„ íš¨ìœ¨ì ì¸ ì‹œìŠ¤í…œì„ êµ¬ì¶•í–ˆìŠµë‹ˆë‹¤. 
# ì´ ì½”ë“œ ì˜ˆì œì—ì„œëŠ” ìˆ˜ë„ ê·¼ì²˜ì˜ ë¬¼ ê³µê¸‰ëŸ‰ê³¼ ê° ë„ì‹œì˜ ìˆ˜ìš”ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³ , 
# truediv ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ë„ì‹œë³„ ë¬¼ ê³µê¸‰ ë¹„ìœ¨ì„ ê³„ì‚°í•©ë‹ˆë‹¤.

import pandas as pd

# ë¬¼ ê³µê¸‰ëŸ‰ (ë‹¨ìœ„: ë§Œ ê°¤ëŸ°)
water_supply = pd.Series({"ë‹ˆë„¤ë² ": 500, "ì•„ìŠˆë¥´": 300}, name="Water Supply")

# ê° ë„ì‹œì˜ ìˆ˜ìš” (ë‹¨ìœ„: ë§Œ ê°¤ëŸ°)
city_demand = pd.Series({"ë‹ˆë„¤ë² ": 450, "ì•„ìŠˆë¥´": 250}, name="City Demand")

# ë„ì‹œë³„ ë¬¼ ê³µê¸‰ ë¹„ìœ¨ ê³„ì‚°
supply_ratio = water_supply.truediv(city_demand, fill_value=1)

print("ğŸ’§ ë„ì‹œë³„ ë¬¼ ê³µê¸‰ ë¹„ìœ¨:")
print(supply_ratio)
```

```python
# ì½”ë“œ ì˜ˆì œ 1: ğŸ›¡ ì•„ì‹œë¦¬ì•„ êµ°ëŒ€ì˜ ë¶€ëŒ€ ë°°ì¹˜ ğŸ›¡
# ì•„ì‹œë¦¬ì•„ ì œêµ­ì€ ê´‘ë²”ìœ„í•œ ì˜í† ë¥¼ í†µì œí•˜ê¸° ìœ„í•´ ì—¬ëŸ¬ êµ°ëŒ€ ë¶€ëŒ€ë¥¼ ë°°ì¹˜í–ˆìŠµë‹ˆë‹¤. 
# ì´ ì½”ë“œ ì˜ˆì œì—ì„œëŠ” ì•„ì‹œë¦¬ì•„ ì œêµ­ì˜ ì „ì²´ êµ°ëŒ€ ìˆ˜ì™€ ê° ë¶€ëŒ€ì— í• ë‹¹í•  ë³‘ì‚¬ ìˆ˜ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³ , 
# floordiv ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ê° ë¶€ëŒ€ì— ë°°ì¹˜ë  ìˆ˜ ìˆëŠ” êµ°ëŒ€ì˜ ìˆ˜ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.

import pandas as pd

# ì•„ì‹œë¦¬ì•„ ì œêµ­ì˜ ì „ì²´ êµ°ëŒ€ ìˆ˜
total_soldiers = pd.Series({"ë³´ë³‘": 12000, "ê¸°ë³‘": 4000}, name="Total Soldiers")

# ê° ë¶€ëŒ€ì— í• ë‹¹í•  ë³‘ì‚¬ ìˆ˜
soldiers_per_unit = pd.Series({"ë³´ë³‘": 300, "ê¸°ë³‘": 200}, name="Soldiers per Unit")

# ê° ë¶€ëŒ€ì— ë°°ì¹˜ë  ìˆ˜ ìˆëŠ” êµ°ëŒ€ì˜ ìˆ˜ ê³„ì‚°
units_assigned = total_soldiers.floordiv(soldiers_per_unit)

print("ğŸ›¡ ê° ë¶€ëŒ€ì— ë°°ì¹˜ë  ìˆ˜ ìˆëŠ” êµ°ëŒ€ì˜ ìˆ˜:")
print(units_assigned)
```

```python
# ì½”ë“œ ì˜ˆì œ 2: ğŸ¹ ì•„ì‹œë¦¬ì•„ ê¶ìˆ˜ ë¶€ëŒ€ì˜ í™”ì‚´ ì†Œëª¨ëŸ‰ ğŸ¹
# ì•„ì‹œë¦¬ì•„ êµ°ëŒ€ëŠ” ì „íˆ¬ì—ì„œ ê¶ìˆ˜ ë¶€ëŒ€ë¥¼ íš¨ê³¼ì ìœ¼ë¡œ í™œìš©í–ˆìŠµë‹ˆë‹¤. 
# ì´ ì½”ë“œ ì˜ˆì œì—ì„œëŠ” ì „íˆ¬ ì¤‘ ê¶ìˆ˜ ë¶€ëŒ€ê°€ ì‚¬ìš©í•  ìˆ˜ ìˆëŠ” ì´ í™”ì‚´ì˜ ìˆ˜ì™€ 
# ê° ê¶ìˆ˜ê°€ ì „íˆ¬ì—ì„œ í‰ê· ì ìœ¼ë¡œ ì‚¬ìš©í•˜ëŠ” í™”ì‚´ì˜ ìˆ˜ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³ , 
# floordiv ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ì „ì²´ ì „íˆ¬ì—ì„œ ê¶ìˆ˜ ë¶€ëŒ€ê°€ ì‚¬ìš©í•  ìˆ˜ ìˆëŠ” í™”ì‚´ì˜ ìˆ˜ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.

import pandas as pd

# ì „íˆ¬ ì¤‘ ê¶ìˆ˜ ë¶€ëŒ€ê°€ ì‚¬ìš©í•  ìˆ˜ ìˆëŠ” ì´ í™”ì‚´ì˜ ìˆ˜
total_arrows = pd.Series({"í™”ì‚´": 100000}, name="Total Arrows")

# ê° ê¶ìˆ˜ê°€ ì „íˆ¬ì—ì„œ í‰ê· ì ìœ¼ë¡œ ì‚¬ìš©í•˜ëŠ” í™”ì‚´ì˜ ìˆ˜
arrows_per_archer = pd.Series({"í™”ì‚´": 20}, name="Arrows per Archer")

# ì „ì²´ ì „íˆ¬ì—ì„œ ê¶ìˆ˜ ë¶€ëŒ€ê°€ ì‚¬ìš©í•  ìˆ˜ ìˆëŠ” í™”ì‚´ì˜ ìˆ˜ ê³„ì‚°
total_battles_supported = total_arrows.floordiv(arrows_per_archer, fill_value=50)

print("ğŸ¹ ì „ì²´ ì „íˆ¬ì—ì„œ ê¶ìˆ˜ ë¶€ëŒ€ê°€ ì‚¬ìš©í•  ìˆ˜ ìˆëŠ” í™”ì‚´ì˜ ìˆ˜:")
print(total_battles_supported)
```

## pandas.Series.mod ë©”ì„œë“œ

```python
# ì½”ë“œ ì˜ˆì œ 1: ğŸ›¡ï¸ ì´ì§‘íŠ¸ êµ°ë‹¨ì˜ ìˆœí™˜ ê·¼ë¬´ ğŸ›¡ï¸
# ì´ì§‘íŠ¸ êµ°ëŒ€ëŠ” ì „ìŸê³¼ ë°©ì–´ë¥¼ ìœ„í•´ ì—¬ëŸ¬ êµ°ë‹¨ì„ ìš´ì˜í–ˆìŠµë‹ˆë‹¤.
# ì´ ì½”ë“œ ì˜ˆì œì—ì„œëŠ” êµ°ë‹¨ë³„ ë³‘ì‚¬ ìˆ˜ì™€ ìˆœí™˜ ê·¼ë¬´ ì¼ì •ì„ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³ , 
# mod ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ìˆœí™˜ ê·¼ë¬´ì— ë”°ë¥¸ ë³‘ì‚¬ë“¤ì˜ ë°°ì¹˜ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.

import pandas as pd

# êµ°ë‹¨ë³„ ë³‘ì‚¬ ìˆ˜
legion_soldiers = pd.Series({"1êµ°ë‹¨": 1000, "2êµ°ë‹¨": 1200, "3êµ°ë‹¨": 1500}, name="Legion Soldiers")

# ìˆœí™˜ ê·¼ë¬´ ì¼ì • (ì¼ ìˆ˜)
rotation_schedule = 7  # í•œ ì£¼ì— í•œ ë²ˆì”© ìˆœí™˜

# ìˆœí™˜ ê·¼ë¬´ì— ë”°ë¥¸ ë³‘ì‚¬ ë°°ì¹˜ ê³„ì‚°
rotation_result = legion_soldiers.mod(rotation_schedule)

print("ğŸ›¡ï¸ ìˆœí™˜ ê·¼ë¬´ì— ë”°ë¥¸ ë³‘ì‚¬ ë°°ì¹˜:")
print(rotation_result)
```

```python
# ì½”ë“œ ì˜ˆì œ 2: ğŸ—¡ï¸ ì „ë¦¬í’ˆì˜ êµ°ë‹¨ë³„ ë¶„ë°° ğŸ—¡ï¸
# ì´ì§‘íŠ¸ êµ°ëŒ€ê°€ ì „íˆ¬ì—ì„œ ìŠ¹ë¦¬í•œ í›„, ì „ë¦¬í’ˆì„ êµ°ë‹¨ë³„ë¡œ ë¶„ë°°í–ˆìŠµë‹ˆë‹¤. 
# ì´ ì½”ë“œ ì˜ˆì œì—ì„œëŠ” ì „ë¦¬í’ˆì˜ ì´ëŸ‰ê³¼ êµ°ë‹¨ë³„ ë³‘ì‚¬ ìˆ˜ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³ , 
# mod ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ì „ë¦¬í’ˆ ë¶„ë°° í›„ ë‚¨ëŠ” ì „ë¦¬í’ˆì˜ ì–‘ì„ ê³„ì‚°í•©ë‹ˆë‹¤.

import pandas as pd

# ì „ë¦¬í’ˆì˜ ì´ëŸ‰
total_spoils = pd.Series({"ê¸ˆ": 700, "ì€": 800}, name="Total Spoils")

# êµ°ë‹¨ë³„ ë³‘ì‚¬ ìˆ˜
legion_soldiers = pd.Series({"ê¸ˆ": 1000, "ì€": 1200}, name="Legion Soldiers")

# ì „ë¦¬í’ˆ ë¶„ë°° í›„ ë‚¨ëŠ” ì–‘ ê³„ì‚°
remaining_spoils = total_spoils.mod(legion_soldiers, fill_value=1)

print("ğŸ—¡ï¸ ì „ë¦¬í’ˆ ë¶„ë°° í›„ ë‚¨ëŠ” ì–‘:")
print(remaining_spoils)
```

## pandas.Series.pow ë©”ì„œë“œ
- Series ê°ì²´ì˜ ê° ìš”ì†Œë¥¼ ì§€ì •ëœ ê°’ìœ¼ë¡œ ê±°ë“­ì œê³±í•˜ëŠ” ì—°ì‚°ì„ ìˆ˜í–‰í•©ë‹ˆë‹¤. ì´ ë©”ì„œë“œëŠ” ìˆ˜í•™ì  ì—°ì‚°ì´ë‚˜ ê¸ˆìœµ ê³„ì‚°ì—ì„œ ë³µë¦¬ ì´ììœ¨ ê³„ì‚°ê³¼ ê°™ì´ ê±°ë“­ì œê³±ì´ í•„ìš”í•  ë•Œ ìœ ìš©í•˜ê²Œ ì‚¬ìš©ë  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
# ì½”ë“œ ì˜ˆì œ 1: ğŸº ì´ì§‘íŠ¸ ë„ìê¸°ì˜ ë‚´êµ¬ì„± í–¥ìƒ ğŸº
# ê³ ëŒ€ ì´ì§‘íŠ¸ì—ì„œ ë„ìê¸°ëŠ” ì¼ìƒ ìƒí™œê³¼ ì˜ë¡€ì—ì„œ ì¤‘ìš”í•œ ì—­í• ì„ í–ˆìŠµë‹ˆë‹¤. 
# ì´ ì½”ë“œ ì˜ˆì œì—ì„œëŠ” ë„ìê¸°ì˜ ì´ˆê¸° ë‚´êµ¬ì„± ê°’ì„ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ì™€
 #ê¸°ìˆ  ë°œì „ì„ í†µí•œ ë‚´êµ¬ì„± í–¥ìƒ ê³„ìˆ˜ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³ , 
 #pow ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ í–¥ìƒëœ ë‚´êµ¬ì„± ê°’ì„ ê³„ì‚°í•©ë‹ˆë‹¤.

import pandas as pd

# ë„ìê¸°ì˜ ì´ˆê¸° ë‚´êµ¬ì„± ê°’
pottery_durability = pd.Series({"ë³‘": 2, "ì ‘ì‹œ": 3, "í•­ì•„ë¦¬": 5}, name="Pottery Durability")

# ê¸°ìˆ  ë°œì „ì— ë”°ë¥¸ ë‚´êµ¬ì„± í–¥ìƒ ê³„ìˆ˜
improvement_factor = 2  # ë‚´êµ¬ì„±ì´ 2ë°°ë¡œ í–¥ìƒë˜ì—ˆë‹¤ê³  ê°€ì •

# í–¥ìƒëœ ë‚´êµ¬ì„± ê°’ ê³„ì‚°
improved_durability = pottery_durability.pow(improvement_factor)

print("ğŸº ì´ì§‘íŠ¸ ë„ìê¸°ì˜ í–¥ìƒëœ ë‚´êµ¬ì„± ê°’:")
print(improved_durability)
```

```python
# ì½”ë“œ ì˜ˆì œ 2: ğŸ—¿ ì´ì§‘íŠ¸ ì¡°ê°ìƒì˜ ë¹„ìœ¨ í™•ëŒ€ ğŸ—¿
# ê³ ëŒ€ ì´ì§‘íŠ¸ì—ì„œ ì¡°ê°ìƒì€ ì¢…êµì , ë¬¸í™”ì  ì¤‘ìš”ì„±ì„ ì§€ë…”ìŠµë‹ˆë‹¤. 
# ì´ ì½”ë“œ ì˜ˆì œì—ì„œëŠ” ì¡°ê°ìƒì˜ ì›ë˜ í¬ê¸° ë¹„ìœ¨ê³¼ ë³µì œ ê³¼ì •ì—ì„œ ì ìš©ëœ í™•ëŒ€ ë¹„ìœ¨ì„ ë‚˜íƒ€ë‚´ëŠ” ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³ , 
# pow ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ í™•ëŒ€ëœ ë¹„ìœ¨ ê°’ì„ ê³„ì‚°í•©ë‹ˆë‹¤.

import pandas as pd

# ì¡°ê°ìƒì˜ ì›ë˜ í¬ê¸° ë¹„ìœ¨
statue_scale = pd.Series({"ì¸ë¬¼ìƒ": 1, "ë™ë¬¼ìƒ": 0.5}, name="Statue Scale")

# ë³µì œ ê³¼ì •ì—ì„œ ì ìš©ëœ í™•ëŒ€ ë¹„ìœ¨
enlargement_factor = pd.Series({"ì¸ë¬¼ìƒ": 2, "ë™ë¬¼ìƒ": 3}, name="Enlargement Factor")

# í™•ëŒ€ëœ ë¹„ìœ¨ ê°’ ê³„ì‚°
enlarged_scale = statue_scale.pow(enlargement_factor)

print("ğŸ—¿ ì´ì§‘íŠ¸ ì¡°ê°ìƒì˜ í™•ëŒ€ëœ ë¹„ìœ¨ ê°’:")
print(enlarged_scale)
```
